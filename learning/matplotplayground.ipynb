```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
x = np.linspace(0, 20, 1000) #Creates a 1d array with 1000 evenly spaced points between 0 and 20
```

```python
y = np.sin(x) + 0.2 * x # arbitrarily add waves
```

```python
plt.plot(x, y);
```

```python
plt.plot(x, y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("my plot"); #semicolon suppresses output if not wanted
```

```python
x = np.random.randn(100, 2) # random 2d array
plt.scatter(x[:,0], x[:,1]); # plot x as the first column and y as the 2nd column
```

```python
x = np.random.randn(200, 2)
x[:50] += 3 # center half the data at (3, 3)
y = np.zeros(200)
y[:50] = 1 # set first half of the labels to 1
plt.scatter(x[:,0], x[:,1], c=y);
```

```python
x = np.random.uniform(-1, 1, (1000, 2)) # uniform distribution between -1 and 1 
x_transformed = np.where(x[:,:] < 0, 0, 1) # transform negative values to 0 and positive values to 1
y = np.where(x_transformed[:,0] != x_transformed[:,1], 1, 0) # xor essentially (first column != second column then value is 1, otherwise 0)
#plt.scatter(y[:,0], y[:,1]);
#plt.scatter(x[:,0], x[:,1]);
plt.scatter(x[:,0], x[:,1], c=y);
```
