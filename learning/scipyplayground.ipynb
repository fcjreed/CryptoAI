```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.signal import convolve2d
from scipy.stats import norm
```

```python
x = np.linspace(-6, 6, 50)
fx = norm.pdf(x, loc=0, scale=1) # location = mean and scale = standard deviation
```

```python
plt.plot(x, fx);
```

```python
Fx = norm.cdf(x, loc=0, scale=1)
plt.plot(x, Fx);
```

```python
logfx = norm.logpdf(x, loc=0, scale=1)
plt.plot(x, logfx);
```

```python
im = Image.open("memetest.jpg")
gray = np.mean(im, axis=2) # convert image to grayscale
filter = np.outer(fx, fx) # calculate a gaussian filter by doing outer product of fx by itself
plt.imshow(filter, cmap="gray");
```

How to blur an image using a filter with convolution

```python
out = convolve2d(gray, filter) # create a 2d convolution to our grayscale image using the filter
plt.subplot(1,2,1) # 1x2 grid of plots and plot first position
plt.imshow(gray, cmap="gray") # plot original grayscale image
plt.subplot(1,2,2) # 1x2 grid of plots and plot second position
plt.imshow(out, cmap="gray"); # plot blurred image that used our convlution
```

Edge detection algorithm

```python
hx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
hy = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
gx = convolve2d(gray, hx)
gy = convolve2d(gray, hy)
val = np.square(gx) + np.square(gy)
g = np.sqrt(val)
plt.imshow(g, cmap="gray");
```
